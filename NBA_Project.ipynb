```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# Got the url from Basketball Reference site

Basketball_Reference_url = 'https://www.basketball-reference.com/leagues/NBA_2024_advanced.html'
response = requests.get(Basketball_Reference_url)

soup = BeautifulSoup(response.text, 'html')
```

```python
print(soup)
```

```python
# Spits out the entire table
table = soup.find_all('table')[0]
```

```python
# th gets out the column headers

table_headers = table.find_all('th')
print(table_headers)
```

```python
table_rows = table.find_all('tr')
```

```python
column_headers = [header.text for header in table_rows[0].find_all(['th', 'tr']) if header.get('data-stat')]
column_headers
```

```python
# Create dataframe

df = pd.DataFrame(columns = column_headers)


for row in table_rows[1:]:
    row_data = [data.text.strip() for data in row.find_all(['th', 'td'])]
    if len(row_data) == len(df.columns):
        df.loc[len(df)] = row_data

df
```
